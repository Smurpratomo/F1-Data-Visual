```python
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions')[1]
full_disp= pd.options.display.max_columns=None
df.head(6)
```

```python
df.columns= df.columns.get_level_values(1)
df = df.loc[:70, :]
df
```

```python
df.drop(4,inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head(6)
```

```python
df.rename(columns={'Chassis':'Constructor'}, inplace=True)
```

```python
df['Country'] = df['Driver'].copy()
df.drop(columns=['Tyres','Poles','Wins','Podiums','F. Laps','Clinched','Margin','% Margin'], inplace=True)
```

```python
df.columns
```

```python
for col in df.columns:
    print(col, ':', df[col].unique())
```

```python
df['Season']= df['Season'].apply(lambda x: x[:-3] if x.endswith('[a]') else x)
df['Driver']= df['Driver'].apply(lambda x: x[:-4])
df['Constructor']= df['Constructor'].apply(lambda x: x[:-3] if x.endswith('[b]') else x)
df['Country']= df['Country'].apply(lambda x: x[-4:])
```

```python
df.head()
```

```python
df['Country']= df['Country'].apply(lambda x: x[1:3])
df['Country']= df['Country'].astype(str).astype(int)
```

```python
country_dict= {18: 'Italy', 19: 'Argentina', 21: 'Italy', 22: 'United Kingdom', 23: 'Australia', 24: 'United States of America', 25: 'United Kingdom', 26: 'United Kingdom', 27: 'United Kingdom', 28: 'New Zealand', 
             29: 'United Kingdom',30: 'Austria', 31: 'Brazil', 32: 'Austria', 33: 'United Kingdom', 34: 'United States of America', 35: 'South Africa', 36: 'Australia', 37: 'Brazil', 38: 'Finland',
             39: 'France', 40: 'Brazil', 41: 'United Kingdom', 42: 'Germany', 43: 'United Kingdom', 44: 'Canada', 46: 'Finland', 47: 'Spain', 48: 'Finland', 49: 'United Kingdom',
             50: 'United Kingdom', 51: 'Germany',52: 'Germany'}
```

```python
df['Country']=df['Country'].replace(country_dict)
```

```python
df['% Points'] = df['% Points'].apply(lambda x: x.split(" ")[0])
```

```python
df['% Points']=df['% Points'].astype(str).astype(float)
df['% Points']=df['% Points'].round(decimals=2)
```

```python
df['Age']=df['Age'].astype(str).astype(int)
```
